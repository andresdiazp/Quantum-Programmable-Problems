# Problema 5

Primero importamos sympy para hacer el cálculo simbólico y numpy para crear la matriz diagonal

```python
import sympy as sp
import numpy as np
```

Definimos las constantes que vamos  utilizar

```python
h = sp.Symbol("ħ")
m = sp.Symbol("m")
w = sp.Symbol("ω")
```

Construimos el operador de ``destrucción`` de ``10x10`` porque nos piden calcular hasta ese número de elementos:

```python
a = sp.Matrix(np.diag([sp.sqrt(i) for i in range(1,10) ], 1))
```

Ahora el operador de ``creación``

```python
a_daga = sp.adjoint(a)
```

# Inciso A)

Primero definimos el operador $\hat{X}$

```python
X = sp.sqrt( (h) / (m*w) ) * (1/sp.sqrt(2)) * (a_daga + a)
```

Ahora solo imprimimos los $\hat{X}^i$ con ``i = 2,3,4,5``

### $\hat{X}^2$

```python
sp.simplify(X**2)
```

### $\hat{X}^3$

```python
sp.simplify(X**3)
```

### $\hat{X}^4$

```python
sp.simplify(X**4)
```

### $\hat{X}^5$

```python
sp.simplify(X**5)
```
